```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rev_month = pd.read_csv('revenue_by_month.csv')
funnel = pd.read_csv('funnel_comparison.csv')
rev_cat = pd.read_csv('revenue_by_category.csv')
```

```python
rev_month.head()
```

```python
rev_month['month'] = pd.to_datetime(rev_month['month'])
```

```python
plt.figure()

plt.plot(rev_month['month'], rev_month['revenue'], marker='o')

plt.title('Revenue by Month')
plt.xlabel('Month')
plt.ylabel('Revenue')

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
funnel.head()
```

```python
import numpy as np

x = np.arange(len(funnel['step']))
w = 0.40

fig, ax = plt.subplots()
ax.bar(x - w/2, funnel['event_based_sessions'], width=w, label='event_based_sessions')
ax.bar(x + w/2, funnel['journey_based_sessions'], width=w, label='journey_based_sessions')

ax.set_xticks(x)
ax.set_xticklabels(funnel['step'], rotation=45)
ax.set_ylabel('Sessions')
ax.set_title('Funnel comparison')
ax.legend(loc=[0.7,1])
```

```python
rev_cat = rev_cat.sort_values("revenue", ascending=False)

fig, ax = plt.subplots()

y_pos = np.arange(len(rev_cat['master_category']))

ax.barh(y_pos, rev_cat['revenue'], align='center')
ax.set_yticks(y_pos, labels=rev_cat['master_category'])
ax.invert_yaxis()
ax.set_xlabel('Revenue')
ax.set_title('Revenue by category')

plt.show()
```
